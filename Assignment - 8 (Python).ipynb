### Q1. In Python, what is the difference between a built-in function and a user-defined function? Provide an example of each.

**Answer:-**  

**1. Built-in function:**  A built-in function is a function that is provided by the Python programming language
as part of its standard library. These functions are pre-defined and readily available
for use without requiring any additional code or module import. Built-in functions are written in C
and are optimized for performance. They cover a wide range of functionalities and provide essential
operations for tasks like string manipulation, mathematical calculations, file handling, etc.

Following example is use to demonstrate built-in function:

```python
x = "Hello Python!"
result = len(x)                           # len() is built-in function to calculate length of desired string
print(f"Length of given string is :" , result)
```

**2. User-defined function:**  A user-defined function is a function created by the programmer to perform a specific task or set of tasks.
These functions are defined using the def keyword followed by a function name, optional parameters,
and a block of code that defines the function's behavior. User-defined functions are modular and allow for code reuse,
abstraction, and better organization of code.

Following example demonstrate user-defined function:


```python
def Add_Numbers(x, y):
    result =  x + y

Add_Numbers(3 , 10)
print(f"Addition of two numbers is : " ,result)
```

### Q2. How can you pass arguments to a function in Python? Explain the difference between positional arguments and keyword arguments.

**Answer:-**  we can pass arguments to a function in multiple ways: positional arguments, keyword arguments, and default arguments.

**1.  Keyword Arguments:**
Keyword arguments allow you to pass arguments to a function by explicitly specifying the parameter name followed by the
corresponding value. This allows you to pass arguments in any order, regardless of their position in the function definition.

Following example demonstrate keyword arguments :    


```python
def Student(name, rollno):
    print("Hello", name, "!")
    print("You have", rollno, "Roll no.")

Student(rollno=20, name="Shruti")
```

**2. Positional Arguments:**
Positional arguments are the most basic way to pass arguments to a function in Python. When you define a function,
you specify the parameters it expects to receive. When you call the function, you pass the corresponding arguments in
the same order as the parameters are defined.

Following example demonstrate positional arguments :

```python
def Student(name, rollno):
    print("Hello", name, "!")
    print("You have", rollno, "Roll no.")

Student("Shruti", 20)
```

**3. Default Arguments:**
 Default arguments allow you to specify a default value for a parameter in a function. If an argument is not provided when calling
the function, the default value will be used.

Following example demonstrate default arguments :

```python
def Student(name, rollno=20):
    print("Hello", name, "!")
    print("You have", rollno, "Roll no.")

Student("Shruti")
```

### Q3. What is the purpose of the return statement in a function? Can a function have multiple return statements? Explain with an example.

**Answer:-**  The **return statement** in a function is to specify the value(s) that the function should output
or "return" back to the caller. It allows the function to compute a result or perform certain actions and provide
the result or information back to the code that called the function.
A function can indeed have multiple return statements. However, when a return statement is encountered, the function
execution immediately stops, and the specified value is returned. This means that only one return statement will be
executed during the function's execution, and the function will terminate at that point.

Following example demonstrate return statement with the help of 'if' , 'else' loop :      

```python
def Result(marks):
    if marks >= 85:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 55:
        return "C"
    elif marks >= 35:
        return "D"
    else:
        return "FAIL!"

student_marks = 92
grade = Result(student_marks)
print(f"The grade for the student is:", grade)
```

### Q4. What are lambda functions in Python? How are they different from regular functions? Provide an example where a lambda function can be useful.

**Answer:-**   **lambda function** is a small anonymous function. It is defined using the lambda keyword, and it can take
any number of arguments, but can only have one expression. Lambda functions are often used in conjunction
with other functions, such as **map()** and **filter()**.

Following example demonstrate use of lambda function :    


```python

double = lambda a: a * 4
double(10)
```

Lambda functions are different from regular functions in a few ways. First, they are anonymous, which means
they do not have a name. Second, they can only have one expression. Third, they are often used in conjunction
with other functions.
example of how a lambda function can be useful. Let's say you have a list of numbers, and you want to double each
number. You could use the map() function to do this:


```python
list_1 = [10, 20, 30, 40, 50]
list_2 = map(lambda a: a * 4, list_1)
print(list_2)
```

 The map() function takes a function and an iterable as arguments. The function is then called with
each item in the iterable, and the results are returned as a new iterable. In this case, the function is the lambda
function double, which doubles each number in the list.

### Q5. How does the concept of "scope" apply to functions in Python? Explain the difference between local scope and global scope.


**Answer:-**   "**Scope"** in Python refers to the region of a program where a variable is accessible.
Variables can only be accessed within their scope. There are three types of scopes in Python:

**1.  Global scope:** This is the scope of the entire program. Variables declared outside of any function are in the global scope.

**2.  Local scope:** This is the scope of a function. Variables declared inside a function are only accessible within that function.

**3.  Built-in scope:** This is the scope of Python's built-in functions and variables. Variables in this scope can be accessed from anywhere in the program.

Local variables are only accessible within the function in which they are declared,
while global variables are accessible from anywhere in the program.

Following examples demonstrate local and global scope of variables :    

```python
# Local scope variable example

def Demo():
    var_1 = 50
    print(var_1)


# Local scope variable example

var_2 = 50
def Demo():
    print(var_2)

Demo()
```

data structure such as a tuple, list, or dictionary.

### Q7. What is the difference between the "pass by value" and "pass by reference" concepts when it comes to function arguments in Python?

**Answer:-**  

**1.  'Pass by Value :**
For immutable objects, such as numbers, strings, and tuples, the value is passed to the function.
Any modifications made to the parameter inside the function do not affect the original object
outside the function.

Following example demonstrate pass by value .

```python
# 'pass by value'

def Pass_Value(x):
    x += 5

a = 500
Pass_Value(a)
print(a)
```

**2.  'Pass by Reference' :**
For mutable objects, such as lists and dictionaries, a reference to the object is passed to the function. Any changes made to the object inside the function can affect the original object outside the function.

Following example demonstrate pass by reference .


```python
# 'pass by reference'

def Pass_Ref(list_1):
    list_1.append(50)

list_2 = [10, 20, 30 ,40]
Pass_Ref(list_2)
print(list_2)
```

### Q8. Create a function that can intake integer or decimal value and do following operations:

a. Logarithmic function (log x)

b. Exponential function (exp(x))

c. Power function with base 2 (2x)

d. Square root


**Answer:-**  

```python
def Operations(x):
    logarithm = calculate_log(x)
    exponential = calculate_exp(x)
    power = calculate_power(x)
    square_root = calculate_square_root(x)
    return logarithm, exponential, power, square_root

def calculate_log(x):
    if x <= 0:
        return None
    result = 0
    while x > 1:
        x /= 10
        result += 1
    return result

def calculate_exp(x):
    result = 1
    for _ in range(int(x)):
        result *= 2.71828
    return result

def calculate_power(x):
    return 2 ** x

def calculate_square_root(x):
    if x < 0:
        return None
    if x == 0:
        return 0
    approx = x / 2
    while True:
        better_approx = (approx + x / approx) / 2
        if abs(approx - better_approx) < 0.0001:
            return better_approx
        approx = better_approx

value = 5.0
results = Operations(value)
print(f"Logarithm Result is : {results[0]}")
print(f"Exponential Result is : {results[1]}")
print(f"Power of 2 Result is : {results[2]}")
print(f"Square Root Result is : {results[3]}")
```

### Q9. Create a function that takes a full name as an argument and returns first name and last name.

**Answer:-**

```python
def Get_Output(full_name):
    names = full_name.split()
    first_name = names[0]
    last_name = names[-1]
    return first_name, last_name

name = "Shruti Saurabh joshi"
first_name, last_name = Get_Output(name)
print("First Name:", first_name)
print("Last Name:", last_name)
```
