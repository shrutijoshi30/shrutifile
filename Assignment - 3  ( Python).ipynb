```python
# In this program , variable is declared outside the function and called outside the function as well.

a = 'Shruti'          # Declaration of the variable outside the function

def MyFunction() :

  print(a)

  global x
  x ='Joshi'         # Declaration of variable as global

MyFunction()

print(x)             # Calling the global variable ouside the function



```
