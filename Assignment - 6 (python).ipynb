###Q.6. What is the operation of this special character ‘\’?

***Answer*** :-  This special character '\' is called as escape character. It is used to introduce special character sequences or escape sequences in strings and characters. The backslash is followed by another character to represent a special meaning or functionality.

Following are some uses of uses of escape character :      

1. The backslash is used to escape characters that would otherwise have a special meaning within a string literal. For example:
'"' represents a double quote character (").
''' represents a single quote character (').
'\ represents a literal backslash character ().
'\n' represents a newline character.
'\t' represents a tab character.
'\r' represents a carriage return character. and many more


2. Line Continuation: The backslash can be used to continue a long line of code onto the next line. This is useful for improving code readability.


3. Raw Strings: By prefixing a string with r or R, the backslashes within the string are treated as literal characters and not as escape characters. This is useful when working with regular expressions or file paths that contain many backslashes.


4. The backslash can be used to represent Unicode characters using their hexadecimal or octal value. For example:
'\uXXXX' represents a Unicode character using its four-digit hexadecimal value.
'\UXXXXXXXX' represents a Unicode character using its eight-digit hexadecimal value.
'\N{...}' represents a Unicode character using its Unicode name.

### Q.8. Explain the mutable and immutable data types with proper explanation & examples.

***Answer*** :- In python ,sometimes we can't change values after creating data types and sometimes we can both the case scenario is known as mutable and immutable data types .

***Mutable data types*** :- Mutable data types are those data types in which we can change or modify values of elements. Any change or updation in the value of element will affect the original values of elements. Following are some mutable data types in python with example :     

1. Dictionaries

2. Sets

3. Lists


```python
list_1 = [10 , 20, 30, 40]
print(list_1)

list_1.append(50)
print(list_1)      # Since list is mutable changed value will affect in the list

list_1[2] = 80
print(list_1)      # Changed value affect in list
```

***Immutable data types*** :- Immutable data types are those whose values cannot be changed or modify once it is assigned. If any operation is performed on an immutable object that appears to modify it, it actually creates a new object with the updated value. Following are some immutable data types in python with example :

1. Tuples

2. Integer , Float , Strings

```python
str_1 = "SHRUTI"
print(str_1)

str_2 = str_1 + "JOSHI"          # String concatination is done which cannot be changed
print(str_2)
```

###Q.9. Write a code to create the given structure using only for loop

        *
       ***
      *****
     *******
    *********


**Answer** :-  Following code demonstrate the above pattern.



```python
rows = 5

for i in range(rows):
    for j in range(rows-i-1):
       print(" " , end = "")

    for a in range(2 * i + 1) :
        print("*" , end = "")

    print()
```

###Q.10. Write a code to create the given structure using while loop.

    |||||||||
     |||||||
      |||||
       |||
        |

**Answer** :-  Following code demonstrate the above pattern.

```python

rows = 5
space = 0
pattern = rows * 2 - 1

while rows > 0:

    i = 0
    while i < space:
        print(" ", end="")
        i += 1

    i = 0
    while i < pattern:
        print("|", end="")
        i += 1

    print()

    space += 1
    pattern -= 2
    rows -= 1
```
