### 9.what are the list concatenation and list replication operations?

**Answer:-**
1.  '+' operator is use to concatinate or merge two lists.

2.  '*' operator is used to replicate or duplicate the list.

Following example demonstrate these two operations on the list

```python
list_1 = ["Orange" , "Apple" , "Banana" , "Mangos" , "Grapes"]
list_2 = ["Rose" , "Lotus" , "Sunflower" , "Marigold" , "Orchid"]

list_3 = list_1 + list_2                     # List concatination

print(f"Concatinated list is :" , list_3)


list_4 = list_2 * 2                          # Replicated list
print(f"Replicated list is :" , list_4)
```

### 10.what is the difference between the list method append() and insert()?

**Answer :-**  

**1. append() :-** This function is used to add new element in the list. By using this method new element is addeed at the last position of the list.

**2. insert() :-** This function is also used to add new element in the list. By using this method new element is added at the specific index position in the list. We have to specify the index value on which new element to be added.

Following example is used to demonstrate these two methods on the list.





```python
list_1 = ["Orange" , "Apple" , "Mangos" , "Grapes"]
list_2 = ["Rose" , "Sunflower" , "Marigold" , "Orchid"]

list_1.append("Cherry")                         # append() function
print(f"After append operation updated list is :" , list_1)

list_2.insert(2 , "Lili")                       # insert() function
print(f"After insert operation updated list is :" , list_2)




```

### 11. What are the two methods for removing items from a list?

**Answers:-**   
1.  **remove()** - This method is used to remove element from the list which occurs first in the list.

2. **pop()** - This method automatically removes last element from the list.

Following example is used to demonstrate these two methods on the list.



```python
list_1 = ["Orange" , "Apple" , "Mangos" , "Grapes"]
list_2 = ["Rose" , "Sunflower" , "Marigold" , "Orchid"]


list_1.pop()                                                 # pop operation
print(f"Updated list after pop operation is :", list_1)


list_2.remove("Sunflower")
print(f"Updated list after remove operation is :", list_2)   # remove operation
```

### 15. How do you get a list value's tuple form? How do you get a tuple value's list form?

**Answer:-** By using following code we can get list value's tuple form and tuple value's list form


```python
# Tuple value's list form

list_1 = ["Pune" ,"Mumbai" , "Nashik"]

tuple_1 = tuple(list_1)

print(type(tuple_1))
print(f"Tuple value's list form :" , tuple_1)


# List value's tuple form

tuple_2 = ["India" ,"USA" , "London"]

list_2 = list(tuple_2)

print(type(list_2))
print(f"List value's tuple form :" , list_2)

```

### 17. How do you distinguish between copy.copy() and copy.deepcopy()?

**Answer:-**  

1. **copy.deepcopy()**: will create a new compound object before inserting copies of the items found in the original into it in a recursive manner.  It means that any changes made to a copy of the object do not reflect in the original object.

2. **copy.copy()**: will creates a new compound object and then references the objects contained in the original within it, which means it constructs a new collection object and then populates it with references to the child objects found in the original.It means that any changes made to a copy of an object do reflect in the original object.
